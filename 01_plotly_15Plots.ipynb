```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# load dataset
df = sns.load_dataset('iris')
```

# 1 scatter plot

```python
fig1 = px.scatter(df, x='sepal_length', y='sepal_width', color='species')
fig1.show()
```

```python
fig1.write_image(".\Output_images\scatter_plot.png", scale=3)
```

# 2 line plot

```python
df_sort = df.sort_values(by=['sepal_length'])
fig1 = px.line(df_sort, x='sepal_length', y='sepal_width', color='species')
fig1.show()
fig1.write_image(".\Output_images\line_plot.png", scale=3)
```

# 3 bar plot

```python
df_bar = df.groupby('species').mean().reset_index()
fig3 = px.bar(df_bar, x='species', y='sepal_length', color='species')
fig3.show()
```

# 4 box plot


```python
fig4 = px.box(df, x='species', y='sepal_length', color='species')
fig4.show()
```

# 5 violon plot

```python
fig5 = px.violin(df, x='species', y='sepal_length', color='species')
fig5.show()
```

# 6 histogram

```python
fig6 = px.histogram(df, x='sepal_length', color='species')
fig6.show()
# save the plot
fig6.write_image(".\Output_images\histogram_plot.png", scale=3)
```

# 7 pie chart

```python
fig7 = px.pie(df, names='species', title='Species Distribution')
fig7.show()
# save the image
fig7.write_image(".\Output_images\pie_chart.png", scale=3)
```

# scatter 3d plot

```python
fig8 = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_length', color='species')
fig8.show()
# save the plot into html
fig8.write_html(".\Output_images\scatter3d_plot.html")
```

# area chart 

```python
df_area = df.sort_values(by=['sepal_length'])
fig9 = px.area(df_area, x='sepal_length', y='sepal_width', color='species')
fig9.show()
fig9.write_image(".\Output_images\area_plot.png", scale=3)
```

# babble chart

```python
fig10 = px.scatter(df, x='sepal_length', y='sepal_width', color='species', size='petal_length')
fig10.show()
fig10.write_image(".\Output_images\scatter_plot_size.png", scale=3)
```

# sumburst chart

```python
df_sunburst = df.groupby(['species', 'petal_width']).mean().reset_index()
fig11 = px.sunburst(df_sunburst, path=['species', 'petal_width'], values='sepal_length')
fig11.show()
fig11.write_image(".\Output_images\sunburst_plot.png", scale=3) 
```

# parallel cordinates plot

```python
df['species_id'] = df['species'].astype('category').cat.codes
fig12 = px.parallel_coordinates(df, color='species_id', labels={'species_id': 'Species'}), color_continuous_scale=px.colors.sequential.Plasma)

fig12.show()
fig12.write_image(".\Output_images\parallel_plot.png", scale=3)
```

# desity contour plot

```python
fig13 = px.density_contour(df, x='sepal_length', y='sepal_width', color='species')
fig13.show()
fig13.write_image(".\Output_images\density_contour_plot.png", scale=3)
```

# 14 ternary plot

```python
fig14 = px.scatter_ternary(df, a='sepal_length', b='sepal_width', c='petal_length', color='species')
fig14.show()
fig14.write_image(".\Output_images\ternary_plot.png", scale=3)
```
